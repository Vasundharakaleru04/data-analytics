```python
import pandas as pd
import matplotlib.pyplot as plt
a=pd.read_csv("class_marks.csv")
a
```

```python
DF=a.sort_values("Total",ascending=True)
DF
```

```python
DF['Q1']=DF["Q1aM4"]+DF["Q1bM6"]
DF['Q2']=DF['Q2aM6']+DF['Q2bM4']
DF['Q3']=DF['Q3aM5']+DF['Q3bM5']
DF['Q4']=DF['Q4aM3']+DF['Q4bM7']
DF['Q6']=DF['Q6aM4']+DF['Q6bM6']
a
```

```python
a=DF.loc[(DF['Total'] >= 15) & (DF['Total'] <= 20)]
a=a.reset_index()
a
```

```python
a1=a.groupby('Q1')['Total']
a1.hist(color = 'blue',figsize=[8,8],grid=False,bins=5)
plt.title("Histogram of students who scored 15-20 Marks")
plt.xlabel("Total Marks obtained")
plt.ylabel("Marks in Q1")
plt.show()
```
