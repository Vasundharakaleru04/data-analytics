```python
data = [
    ("Murder", "For_gain", 28, 2, 3),
    ("Murder", "Property_Dispute", 20, 1, 2),
    ("Cyber_Crime", "Job_Fraud", 65, 3, 51),
    ("Cyber_Crime", "OTP_Fraud", 93, 3, 77),
    ("POCSO_Act", "Rape", 2446, 229, 153)
]
```

```python
b
```

```python
b = pd.DataFrame(data, columns=["Category", "Subcategory", "Current_Year", "Prev_Month", "Current_Month"])
```

```python
b
```

```python
print(b.head())
```

```python
print("Missing Values:\n", b.isnull().sum())

# Convert numeric columns to integers (if needed)
b["Current_Year"] = b["Current_Year"].astype(int)
b["Prev_Month"] = b["Prev_Month"].astype(int)
b["Current_Month"] = b["Current_Month"].astype(int)

print("\nData cleaned successfully.")
```

```python
category_totals = b.groupby("Category")["Current_Year"].sum()
print("\nTotal Crimes by Category:\n", category_totals)

# Percentage change from previous month
b["Percent_Change"] = ((b["Current_Month"] - b["Prev_Month"]) / b["Prev_Month"]) * 100
print("\nMonthly Crime Trend:\n", b[["Category", "Subcategory", "Percent_Change"]])
```

```python
b = b.apply(pd.to_numeric, errors='coerce')  # Convert to numeric, replace non-numeric with NaN
```
